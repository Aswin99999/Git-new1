```python
# I want to see Region wise Total salses, Region ---> index or column, values = sales, aggfunc = sum

df.pivot_table(index='Region', values='Sales', aggfunc=sum)
```

```python
df.pivot_table(columns='Region',values='Sales',aggfunc=sum)
```

```python
# I want to have region wise and product wise total sales, index --> Region, column = 'Type', values = Sales, aggfunc = sum

df.pivot_table(index='Region',columns='Type',values='Sales',aggfunc=sum)
```

```python
df.pivot_table(index='Region',columns='Type',values='Sales',aggfunc=sum,fill_value=0)
```

```python
df.pivot_table(index=['Region','Type'],values='Sales',aggfunc=sum)
```

```python
df.pivot_table(columns=['Region','Type'],values='Sales',aggfunc=sum,fill_value=0)
```

#### Region wise average unit sold and average sales by each salesperson

```python
df.pivot_table(index=['Region','Salesperson'],values=['Units','Sales'],aggfunc='mean')
```

### Region wise average unit sold and average sales by each salesperson fetch the info for different Types of the product and Year wise 

```python
df.pivot_table(index=['Year','Salesperson','Region'],values=['Units','Sales'],columns=['Type'],fill_value=0,aggfunc='mean')
```
