#### capping

```python
import numpy as np
```

```python
df['New_Age'] = np.where( df['Age']<0 , 0, np.where(df['Age']>60,60,df['Age'])    )
```

```python
df
```

```python
sns.boxplot(y='New_Age',data=df)
```

```python
df['new_age'] = df.Age.clip(lower=0,upper=60)
```

```python
df
```

```python
sns.boxplot(y='new_age',data=df)
```
